- Круглое или овальное лицо (1) 
Хороший хозяин, любит комфорт и роскошь. Дружелюбный, обладает развитой способностью суждения и интуицией, имеющий определенные ценности и сильное чувство справедливости. Способный, усердно работающий в бизнесе, но иной раз предающийся лени и самооправданиям. Обычно такая форма говорит о беззаботном, веселом, добродушном человеке, который любит проявления комфорта в жизни: физический комфорт, ощущения от прикосновения к вещам, хорошая пища. Люди с круглыми лицами, как правило, хорошие любовники. Эта форма лица соответствует эндоморфному типу. 

- Треугольное или сердцевидное лицо (2)
Сообразительный, интеллектуальный, обладает хорошей памятью, возможны перепады настроения. Пытается мечты воплотить в жизнь, может быть непрактичным. Чувствительный, склонный к самонаблюдению, легко теряющий интерес и начинающий скучать, ему постоянно нужно чувство цели. Эта форма отражает чувствительную личность, особенно если человек худощав и не имеет второго подбородка. Чем туже натянута кожа, тем чувствительнее человек. Индивид такого типа может быть интровертом, при этом он чрезвычайно умен. Люди с треугольными лицами могут быть мечтательны, а могут – коварны, могут быть ревнивы и не хранить верность своим любимым. Эта форма соответствует эктоморфному типу. 

- Квадратное или прямоугольное лицо (3) 
Природный лидер, подвижный, добросовестный, всегда стремится быть активным. Интеллигентен, практичен, обладает разнообразными навыками. Физически силен, а также обладает сильной волей, мнение высказывает жестко и решительно. Эта форма отражает суровую, непреклонную личность, возможно, довольно вспыльчивую. Люди с такой формой лица – возможные лидеры и воины; обычно они откровенны как в бизнесе, так и в любовных делах и стремятся к успеху. Квадратное лицо соответствует мезоморфному типу. ![alt text](https://www.starfate.ru/assets/images/ezo/30_05_11/hiromantija/1/image002.jpg)

- Зона 1. Верхняя зона Лоб указывает на интеллектуальные способности. Хорошо развитый лоб обычно свидетельствует об уме человека. В идеальном случае уши должны находиться на большом расстоянии друг от друга, а лоб – быть длинным и высоким. 

- Зона 2. Средняя зона Если средняя зона лица длинная – при наличии высокого лба и длинной нижней зоны, – человек обладает благородными качествами. Если средняя зона слишком длинна по сравнению с другими зонами, человек может быть высокомерным, но наделенным духом предприимчивости. Если средняя зона находится в идеальной (равной) пропорции с другими зонами, этот человек, вероятно, долго проживет. 

- Зона 3. Нижняя зона Если нижняя зона длиннее других зон и при этом сильная, широкая и, возможно, мясистая, это указывает на успех и процветание в жизни. Если подбородок длинный и заостренный, человеку, вероятно, необходимо изменить свой взгляд на жизнь и научиться ладить с окружающими. Людям с нижней зоной такого типа следует обратить внимание на науку или философию, а также гуманитарные дисциплины, где они могут найти более одухотворенное знание. Длинная и широкая нижняя зона указывает на человека, уделяющего большое внимание физическим качествам. Слишком узкая нижняя зона свидетельствует, что человека больше волнует внутренний мир, нежели физическая сторона, и в вопросах любви ему свойственно богатое воображение. Каждая часть любой из трех зон лица имеет свои сигналы и признаки, свое значение и свой смысл для эмоциональной жизни. Искусство чтения по лицу не ограничивается только формой лица и его зонами, для более полной картины конечно требуется учитывать форму глаз, бровей, носа и губ, для этого вам понадобится дополнительная литература.
```
# Подгружаем таблицу
data = pd.read_excel('05_12_2019_Fiziognomika.xlsx')

# Удаляем пробелы и лишние символы
data = data.replace(r'\s+','',regex=True)

# Доавляем имена картинок
data['filenames'] = pictures_names_list

# удаляем лишние столбцы
data = data.drop(['Unnamed: 0',	'Unnamed: 1'], axis =1)

# Превращаем описание в лист
data['labels'] = data['labels'].apply(lambda x:x.split(","))
data.head(3)
```

## Создание нейронной сети

### Создаем генераторы

```
# Используем библиотеку keras_preprocessing.image, и делаем генераторы датасета

datagen = ImageDataGenerator(rescale = 1./255.)
test_datagen = ImageDataGenerator(rescale=1./255.)

# Train  генератор
# Указываем количество картинок, директорию, Х и У, батч, сид, перемешивание, класс - categorical, названия классов, размер изображения
train_generator = datagen.flow_from_dataframe(\
                  dataframe = data[:440],\
                  directory = '/content/drive/My Drive/Colab Notebooks/Neural university/Neural networks/00_Diploma/diploma_pic/',\
                  x_col = 'filenames',\
                  y_col = 'labels',\
                  batch_size = 32,\
                  seed = 42,\
                  shuffle = True,\
                  class_mode='categorical',\
                  classes = ['тиран', 'духовный',	'Ответственный', 'узкомыслящий',\
                             'хитрый', 'семейный',	'самоуверенный',	'мыслящий широко',\
                             'чувствительный', 'материальный', 'добродушный', 'артистичный','мягкий'],\
                  target_size = (100,100))

# Valid  генератор
valid_generator=test_datagen.flow_from_dataframe(\
                dataframe=data[440:475],\
                directory='/content/drive/My Drive/Colab Notebooks/Neural university/Neural networks/00_Diploma/diploma_pic/',\
                x_col="filenames",\
                y_col="labels",\
                batch_size=32,\
                seed=42,\
                shuffle=True,\
                class_mode="categorical",\
                classes = ['тиран', 'духовный',	'Ответственный', 'узкомыслящий',\
                             'хитрый', 'семейный',	'самоуверенный',	'мыслящий широко',\
                             'чувствительный', 'материальный', 'добродушный', 'артистичный','мягкий'],\
                target_size=(100,100))
# Test  генератор
test_generator=test_datagen.flow_from_dataframe(\
                dataframe=data[475:],\
                directory='/content/drive/My Drive/Colab Notebooks/Neural university/Neural networks/00_Diploma/diploma_pic/',\
                x_col="filenames",\
                batch_size=1,\
                seed=42,\
                shuffle=True,\
                class_mode=None,\
                target_size=(100,100))
```

### Архитектура

```
# Оптимальная архитектура нейронки
model = Sequential()

model.add(Conv2D(32, (3, 3), padding='same',
                 input_shape=(100,100,3)))
model.add(Activation('relu'))

model.add(Conv2D(32, (3, 3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))

model.add(Conv2D(64, (3, 3), padding='same'))
model.add(Activation('relu'))

model.add(Conv2D(64, (3, 3)))
model.add(Activation('relu'))
model.add(MaxPooling2D(pool_size=(2, 2)))
model.add(Dropout(0.25))
model.add(Flatten())

model.add(Dense(512))
model.add(Activation('relu'))
model.add(Dropout(0.5))
model.add(Dense(13, activation='sigmoid'))
model.compile(optimizers.rmsprop(lr=0.0001, decay=1e-6),\
              loss="binary_crossentropy",metrics=["accuracy"])
```

```
print(model.summary())
```

### обучение

```
STEP_SIZE_TRAIN=train_generator.n//train_generator.batch_size # шаг обучения на обучающем генераторе
STEP_SIZE_VALID=valid_generator.n//valid_generator.batch_size # шаг обучения на валидационном  генераторе
STEP_SIZE_TEST=test_generator.n//test_generator.batch_size # шаг обучения на тестовом генераторе

#Обучаем нейронку
history = model.fit_generator(generator=train_generator,
                    steps_per_epoch=STEP_SIZE_TRAIN,
                    validation_data=valid_generator,
                    validation_steps=STEP_SIZE_VALID,
                    epochs=200)
#РИсуем график
plt.plot(history.history['acc'], 
         label='Доля верных ответов на обучающем наборе')
plt.plot(history.history['val_acc'], 
         label='Доля верных ответов на проверочном наборе')
plt.xlabel('Эпоха обучения')
plt.ylabel('Доля верных ответов')
plt.legend()
plt.show()
```

### Тестирование

```
test_generator.reset() # важно сделать сброс генератора

pred=model.predict_generator(test_generator,steps=STEP_SIZE_TEST,verbose=1) # запускаем предикт
```

```
pred_bool = (pred >0.5) # создаем массив с предиктом, указываем вероятность пренадлежности к классу больше 0,5
```

```
type(pred_bool) # смотрим тип данных
```

### сохраняем

```
predictions=[]# пустой список для предсказаний (для создания датафрейма)

labels = train_generator.class_indices
labels = dict((v,k) for k,v in labels.items())

for row in pred_bool: # проходим по массиву с предсказаниями
    l=[] # создаем пустой список для лейблов
    for index,cls in enumerate(row):
        if cls:
            l.append(labels[index]) # добавляем лейблы
    predictions.append(",".join(l)) 

filenames=test_generator.filenames # обращаемся к именам картинок

# создаем датафрейм и сохраняем
results=pd.DataFrame({"Filename":filenames,
                      "Predictions":predictions})
results.to_csv("results.csv",index=False)
```
